```python
import lighteval
from lighteval.logging.evaluation_tracker import EvaluationTracker
from lighteval.models.vllm.vllm_model import VLLMModelConfig
from lighteval.pipeline import ParallelismManager, Pipeline, PipelineParameters
#from lighteval.utils.imports import is_package_available
```

```python
evaluation_tracker = EvaluationTracker(output_dir="./results", save_details=True, push_to_hub=False)
```

```python
pipeline_params = PipelineParameters(
        launcher_type=ParallelismManager.ACCELERATE,
        custom_tasks_directory=None,  # Set to path if using custom tasks
        # Remove the parameter below once your configuration is tested
        max_samples=10
    )
```

```python
model_config = VLLMModelConfig(model_name="HuggingFaceH4/zephyr-7b-beta", dtype="float16")
#model_config = VLLMModelConfig(model_name="stabilityai/stablelm-zephyr-3b", dtype="bfloat16")
```

```python
# task = "gsm8k|5"
task = "lighteval|gsm8k|5|1" # suite|task|few_shot|truncate_few_shots
```

```python
pipeline = Pipeline(tasks=task, pipeline_parameters=pipeline_params, evaluation_tracker=evaluation_tracker, 
    model_config=model_config)
```

```python
pipeline.evaluate()
pipeline.save_and_push_results()
pipeline.show_results()
```
