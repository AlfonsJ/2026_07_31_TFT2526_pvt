```python
model = "ojus1/Qwen3-0.6B-Instruct"
```

```python
from trl import SFTTrainer
from datasets import load_dataset
trainer = SFTTrainer(model=model, train_dataset=load_dataset("trl-lib/Capybara", split="train"))
trainer.train()
```

```python
from trl import GRPOTrainer
from datasets import load_dataset
# Define a simple reward function (count unique chars as example)
def reward_function(completions, **kwargs):
    return [len(set(completion.lower())) for completion in completions]

trainer = GRPOTrainer(
    model=model,  # Start from SFT model
    train_dataset=load_dataset("trl-lib/tldr", split="train"),
    reward_funcs=reward_function,
)
trainer.train()
```

```python
from trl import DPOTrainer
from datasets import load_dataset

model = "ojus1/Qwen3-0.6B-Instruct"
trainer = DPOTrainer(model=model, ref_model=model, train_dataset=load_dataset("trl-lib/ultrafeedback_binarized", split="train"))
trainer.train()
```
